```python
import random, sys

print('ROCK, PAPER, SCISSORS')

# These variables keep track of the number of wins, losses, and ties.
wins = 0
losses = 0
ties = 0

while True: # The main game loop.
    print('%s Wins, %s Losses, %s Ties' % (wins, losses, ties))
    while True: # The player input loop.
        print('Enter your move: (r)ock (p)aper (s)cissors or (q)uit')
        playerMove = input()
        if playerMove == 'q':
            sys.exit() # Quit the program.
        if playerMove == 'r' or playerMove == 'p' or playerMove == 's':
            break # Break out of the player input loop.
        print('Type one of r, p, s, or q.')

    # Display what the player chose:
    if playerMove == 'r':
        print('ROCK versus...')
    elif playerMove == 'p':
        print('PAPER versus...')
    elif playerMove == 's':
        print('SCISSORS versus...')

    # Display what the computer chose:
    randomNumber = random.randint(1, 3)
    if randomNumber == 1:
        computerMove = 'r'
        print('ROCK')
    elif randomNumber == 2:
        computerMove = 'p'
        print('PAPER')
    elif randomNumber == 3:
        computerMove = 's'
        print('SCISSORS')

    # Display and record the win/loss/tie:
    if playerMove == computerMove:
        print('It is a tie!')
        ties = ties + 1
    elif playerMove == 'r' and computerMove == 's':
        print('You win!')
        wins = wins + 1
    elif playerMove == 'p' and computerMove == 'r':
        print('You win!')
        wins = wins + 1
    elif playerMove == 's' and computerMove == 'p':
        print('You win!')
        wins = wins + 1
    elif playerMove == 'r' and computerMove == 'p':
        print('You lose!')
        losses = losses + 1
    elif playerMove == 'p' and computerMove == 's':
        print('You lose!')
        losses = losses + 1
    elif playerMove == 's' and computerMove == 'r':
        print('You lose!')
        losses = losses + 1
```

"t" - Text - Default value. Text mode

# XML parsing

* XML stands for "Extensible Markup Language"
* XML documents have sections called elements, which are defined by a beginning and an ending tag. A tag is a markup construct that begins with < and ends with >. The characters between the start-tag and end-tag, if there are any, are the element's content. Elements can contain markup, including other elements, which are called "child elements."
* The largest, top-level element is called the root, which contains all other elements.
* Attributes are namevalue pair that exist within a start-tag or empty-element tag. An XML attribute can only have a single value and each attribute can appear at most once on each element.

```python
tree = ET.parse('books.xml')
root = tree.getroot()
```

* Now that you have initialized the tree, you should look at the XML and print out values in order to understand how the tree is structured.

* Every part of a tree (root included) has a tag that describes the element. In addition, as you have seen in the introduction, elements might have attributes, which are additional descriptors used especially for repeated tag usage.

```python
root.tag
```

At the top level, you see that this XML is rooted in the **catalog** tag.

```python
root.attrib
```

So, the root has no attributes.

### For loops

You can easily iterate over subelements (commonly called "children") in the root by using a simple "for" loop.

```python
for child in root:
    print(child.tag)
    print (child.attrib)
```

Now you know that the children of the root **catalog** are all **book**. To designate the **book**, the XML uses the attribute **id**

```python
[elem.tag for elem in root.iter()]
```

Typically it is helpful to know all the elements in the entire tree. One useful function for doing that is root.iter(). You can put this function into a "for" loop and it will iterate over the entire tree.

```python
for x in root.iter('book'):
    print(x.attrib)
```

Many times elements will not have attributes, they will only have text content. Using the attribute .text, you can print out this content.

Now, print out all the descriptions of the books.


```python
for description in root.iter('description'):
    print (description.text)
```

##### Understanding XPath is critically important to scanning and populating XMLs. ElementTree has a .findall() function that will traverse the immediate children of the referenced element. You can use XPath expressions to specify more useful searches.

```python
for y in root.findall("./book/author"):
    print(y.text)
```

```python
for book in root.findall("./book"):
    price = float(book.find('price').text)
    if price>30:
        title = book.find('title').text
        print(f"{title} is > 30 its price is {price}")
```

XPath is a query language used to search through an XML quickly and easily. XPath stands for XML Path Language and uses, as the name suggests, a "path-like" syntax to identify and navigate nodes in an XML document.

```python
for book in root.find("./book[publish_date='2000-11-17']"):
    print(book.text)
```

*  <em>Element.iter( )</em>: <p>Element has some useful methods that help iterate recursively over all the sub-tree below it (its children, their children, and so on).</p>

* <em>Element.findall( ):</em>: <p>Finds only elements with a tag which are direct children of the current element.</p>

* <em>Element.find( ):</em> <p>finds the first child with a particular tag</p>

* <em>Element.text( ):</em> <p>accesses the elements text content.</p>

* <em>Element.get( ):</em> <p>accesses the elements attributes</p>

```python
for details in root.findall('book'):
    genre = details.find('genre').text
    price = details.find('price').text
    print(f"For the genre:{genre} the price is {price}")
```

### Modifying an XML File


<p>ElementTree provides a simple way to build XML documents and write them to files. The <strong>ElementTree.write( )</strong> method serves this purpose.
</p>

<p> Element object may be manipulated by directly changing its fields (such as <strong>Element.text</strong>), adding and modifying attributes (<strong> Element.set()</strong> method), as well as adding new children (for example with <strong>Element.append()</strong>).</p>

Lets say we want to add one to each countrys rank, and add an **updated** attribute to the rank element:

>Let us imagine we want to increase the price by 1$


```python
tree = ET.parse('books.xml')  # Replace with the correct path
root = tree.getroot()

for newPrice in root.iter('price'):
    updatedPrice = float(newPrice.text) + 1  # Increase price by 1
    newPrice.text = str(updatedPrice)        # Update the element's text
    newPrice.set('updated', 'yes')           # Add/update an attribute

tree.write('output.xml')
```
